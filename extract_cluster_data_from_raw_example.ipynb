```python
import xarray as xr
import numpy as np
import glob
```

```python
# Load one of the raw data files
datapath = '/lus/scratch/shao/data/NEP36_extremes/'
file = f"{datapath}/NEP36-CanOE_1d_benthic_shelf_20200217-20200217.nc"
```

```python
# Subset the data file in the same way that the cluster data was (limited to a smaller domain)
x_slice = slice(215,713)
y_slice = slice(180,750)

ds = xr.open_dataset(file)
ds = ds.isel(x=x_slice, y=y_slice)
```

```python
# Load the climatology with the cluster labels for every x, y point
cluster_clim_ds = xr.open_dataset(f"{datapath}/processed/clim/clim_with_clusters.nc")
ds["cluster"] = cluster_clim_ds["cluster"]
```

```python
# Option 1: Create groups to only do analysis on a given one
ds_clusters = ds.groupby("cluster")
cluster_labels = list(ds_clusters.groups.keys())
ds_clusters[cluster_labels[0]].mean()
```
